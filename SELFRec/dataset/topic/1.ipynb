```python
import torch
import numpy as np
user_emb = torch.load('user_emb2.pt')
item_emb = torch.load('item_emb2.pt')
import json
u = '211'
k=75
# 从文件中加载字典
with open('user.json', 'r') as f:
    user = json.load(f)

with open('item.json', 'r') as f:
    item = json.load(f)

```

```python
user_emb[0]
```

```python
def get_user_id(u):
        if u in user:
            return user[u]
def get_item_id(i):
        if i in item:
            return item[i]
```

```python

def predict(u):
    u = get_user_id(u)
    score = torch.matmul(user_emb[u], item_emb.transpose(0, 1))
    return score
```

```python
u_i = {}

with open("data.txt","r") as f:
    for line in f:
        line = line.split(" ")
        if line[0] not in u_i.keys():
            u_i[line[0]]=[line[1]]

        else:
            u_i[line[0]].append(line[1])

u_i

```

```python
score = predict(u)
print(score)
```

```python
values, indices = torch.topk(score,k=k)
indices = indices.cpu().numpy()
print(indices)
```

```python

def get_item(id):
    for key, value in item.items():
        if value == id:
            return key
        
```

```python
rec_item = []
for i in indices:
    rec_item.append(get_item(i))
print(rec_item)
```

```python
temp = 0
print(len(u_i[u]))
for i in u_i[u]:
    if i in rec_item:
        temp += 1

print(temp)

```
